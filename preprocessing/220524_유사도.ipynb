```python
pattern1 = '--- p.[0-9]+~[0-9]+'
pattern2 = '--- p.[0-9]+'
pattern3 = 'nan'
pattern4 = r'\r'
pattern5 = r'\n'


sentence_list = []
for i, v in datadf.iterrows():
    sentence = str(v['title']) + str(v['0']) + str(v['1']) + str(v['2']) + str(v['3']) + str(v['4']) + str(v['5'])
    sentence = re.sub(pattern1, '', sentence)
    sentence = re.sub(pattern2, '', sentence)
    sentence = re.sub(pattern3, '', sentence)
    sentence = re.sub(pattern4, '', sentence)
    sentence = re.sub(pattern5, '', sentence)

    sentence_list.append(sentence)
```

### Bag of Words (BoW)

```python
from konlpy.tag import Okt

okt = Okt()
def build_bag_of_words(document):
    # 온점 제거 및 형태소 분석
    document = document.replace('.', '')
    tokenized_document = okt.morphs(document)

    word_to_index = {}
    bow = []

    for word in tokenized_document:  
        if word not in word_to_index.keys():
            word_to_index[word] = len(word_to_index)  
            # BoW에 전부 기본값 1을 넣는다.
            bow.insert(len(word_to_index) - 1, 1)
        else:
            # 재등장하는 단어의 인덱스
            index = word_to_index.get(word)
            # 재등장한 단어는 해당하는 인덱스의 위치에 1을 더한다.
            bow[index] = bow[index] + 1

    return word_to_index, bow
```

```python
doc1 = "정부가 발표하는 물가상승률과 소비자가 느끼는 물가상승률은 다르다."
vocab, bow = build_bag_of_words(doc1)
print('vocabulary: ', vocab)
print('bag of words vector: ', bow)
```

```python
doc2 = '소비자는 주로 소비하는 상품을 기준으로 물가상승률을 느낀다.'

vocab, bow = build_bag_of_words(doc2)
print('vocabulary :', vocab)
print('bag of words vector :', bow)
```

```python
doc3 = doc1 + ' ' + doc2
vocab, bow = build_bag_of_words(doc3)
print('vocabulary :', vocab)
print('bag of words vector :', bow)
```

```python
# 총 문서의 수
N = len(sentence_list) 

def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in sentence_list:
        df += t in doc
    return np.log(N/(df+1))

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
result = []

# 각 문서에 대해서 아래 연산을 반복
for i in range(N):
    result.append([])
    d = sentence_list[i]
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
```
